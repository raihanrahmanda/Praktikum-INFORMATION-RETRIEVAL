```python
#NO1

import os
def tokenisasi(text):
    tokens = text.split(" ")
    return tokens
def stemming(text):
    from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
    # create stemmer
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    # stemming process
    output = stemmer.stem(text)
    return output
def stemming_sentence(text):
    output = ""
    for token in tokenisasi(text):
        output = output + stemming(token) + " "
    return output[:-1]

berita_folder = "C:/Users/Afi/Downloads/berita~/berita"

doc_dict_raw = {}
for filename in os.listdir(berita_folder):
        if filename.endswith(".txt"):
            with open(os.path.join(berita_folder, filename), "r", encoding="utf-8") as file:
                doc_dict_raw[filename] = file.read()

doc_dict = {}
for doc_id,doc in doc_dict_raw.items():
    doc_dict[doc_id] = stemming_sentence(doc)
print(doc_dict)
```

```python
vocab = []
inverted_index = {}
for doc_id,doc in doc_dict.items():
    for token in tokenisasi(doc):
        print(token)
        if token not in vocab:
            vocab.append(token)
            inverted_index[token] = []
        if token in inverted_index:
            if doc_id not in inverted_index[token]:
                inverted_index[token].append(doc_id)

print(vocab)
print(inverted_index)
```

```python
def termFrequency(vocab, query):
    tf_query = {}
    for word in vocab:
        tf_query[word] = query.count(word)
    return tf_query

import numpy as np
def inverseDocFre(vocab,doc_fre,length):
    idf= {}
    for word in vocab:
        idf[word] = idf[word] = 1 + np.log((length + 1) / (doc_fre[word]+1))
    return idf

def wordDocFre(vocab, doc_dict):
    df = {}
    for word in vocab:
        frq = 0
        for doc in doc_dict.values():
            if word in tokenisasi(doc):
                frq = frq + 1
        df[word] = frq
    return df

def tfidf(vocab,tf,idf_scr):
    tf_idf_scr = {}
    for doc_id in doc_dict.keys():
        tf_idf_scr[doc_id] = {}
    for word in vocab:
        for doc_id,doc in doc_dict.items():
            tf_idf_scr[doc_id][word] = tf[doc_id][word] * idf_scr[word]
    return tf_idf_scr

query = "vaksin corona jakarta"

tf_query = termFrequency(vocab, query)
idf_scr = inverseDocFre(vocab, wordDocFre(vocab, doc_dict), len(doc_dict))


tfidf_query = {}
for word in vocab:
    tfidf_query[word] = tf_query[word] * idf_scr[word]

    # Term - Query Matrix
TQ = np.zeros((len(vocab), 1)) #hanya 1 query
for word in vocab:
    ind1 = vocab.index(word)
    TQ[ind1][0] = tf_query[word]*idf_scr[word]
print(TQ)

def termFrequencyInDoc(vocab, doc):
    tf_doc = {}
    for word in vocab:
        tf_doc[word] = doc.count(word)
    return tf_doc

tfidf_docs = {}
for doc_id, doc in doc_dict.items():
    tf_doc = termFrequencyInDoc(vocab, doc)
    tfidf_doc = {}
    for word in vocab:
        tfidf_doc[word] = tf_doc[word] * idf_scr[word]
    tfidf_docs[doc_id] = tfidf_doc

TD = np.zeros((len(vocab), len(doc_dict)))
for i, doc_id in enumerate(doc_dict.keys()):
    for j, word in enumerate(vocab):
        TD[j][i] = tfidf_docs[doc_id][word]
```

```python
import math
def cosine_sim(vec1, vec2):
    vec1 = list(vec1)
    vec2 = list(vec2)
    dot_prod = 0
    for i, v in enumerate(vec1):
        dot_prod += v * vec2[i]
    mag_1 = math.sqrt(sum([x**2 for x in vec1]))
    mag_2 = math.sqrt(sum([x**2 for x in vec2]))
    return dot_prod / (mag_1 * mag_2)

print(cosine_sim(TQ[:, 0], TD[:, 0])) #query & doc1
print(cosine_sim(TQ[:, 0], TD[:, 1])) #query & doc2
print(cosine_sim(TQ[:, 0], TD[:, 2])) #query & doc3
print(cosine_sim(TQ[:, 0], TD[:, 3])) #query & doc4
print(cosine_sim(TQ[:, 0], TD[:, 4])) #query & doc5
```

```python
from collections import OrderedDict
def exact_top_k(doc_dict, TD, q, k):
    relevance_scores = {}
    i = 0
    for doc_id in doc_dict.keys():
        relevance_scores[doc_id] = cosine_sim(q, TD[:, i])
        i = i + 1
    sorted_value = OrderedDict(sorted(relevance_scores.items(), key=lambda x: x[1], reverse = True))
    top_k = {j: sorted_value[j] for j in list(sorted_value)[:k]}
    return top_k
```

```python
top_3 = exact_top_k(doc_dict, TD, TQ[:, 0], 3)
print(top_3)
```

```python
#NO2

def index_elim_simple(query, doc_dict):
    remove_list =[]
    for doc_id,doc in doc_dict.items():
        n = 0
        for word in tokenisasi(query):
            if stemming(word) in doc:
                n = n+1

        if n==0:
            remove_list.append(doc_id)

    for key in remove_list:
        del doc_dict[key]
    return doc_dict
```

```python
doc_dict_skor = index_elim_simple(query, doc_dict)
print(doc_dict_skor)
```

```python
def elim_query(query, idf_dict, idf_score):
    for term in tokenisasi(query):
        if idf_dict[stemming(term)]<idf_score:
            query = query.replace(term+" ", "")
            query = query.replace(term, "")
    return query

query = "vaksin corona jakarta"
query = elim_query(query, idf_scr, 1.5)
print(query)
```
